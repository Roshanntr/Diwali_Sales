```python
ax=sns.countplot(x='Gender',data=df,palette='bright') 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount', data=sales,palette='husl')
```

```python
sns.countplot(x='Age Group',data=df,palette='husl')
sns.set(rc={'figure.figsize':(7,5)})
```

```python
age_wise=sns.countplot(x='Age Group',data=df,hue='Gender') 
sns.set(rc={'figure.figsize':(7,5)})
for bars in age_wise.containers:
    age_wise.bar_label(bars)
```

```python
sale=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Age Group',y='Amount', data=sale,palette='deep')
```

From the above graph, we can see that, most of the buyers are of age group -26 to 35

```python
states=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Orders',data=states,palette='husl')
```

Most of the orders are from Uttar Pradesh and Maharashtra

```python
stat=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Amount',data=stat,palette='muted')
```

Most of the order and sales are from Uttar pradesh,Maharashtra and Karnataka

```python
ax=sns.countplot(data=df,x='Marital_Status',palette='deep')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
stat=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status',y='Amount',data=stat,hue='Gender',palette='muted')
```

From the above plot, Married women have higher purchasing power.

```python
ax=sns.countplot(data=df,x='Occupation',palette='husl')
sns.set(rc={'figure.figsize':(20,10)})

```

```python
stat=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=stat,palette='deep')
```

Most of the buyers are from IT sector,Healthcare and Aviation.

```python
ax=sns.countplot(data=df,x='Product_Category',palette='husl')
sns.set(rc={'figure.figsize':(25,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
stat=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=stat,palette='deep')
```

Most sold product is Food, Clothing & Apparel

```python
stat=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Amount',data=stat,palette='husl')
```
